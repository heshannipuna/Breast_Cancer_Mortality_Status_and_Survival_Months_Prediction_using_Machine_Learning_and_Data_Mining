```python
#IQR outlier detection function
def find_outliers_IQR(data):

  q1=data.quantile(0.25)  #Q1 - 25th percentile
  q3=data.quantile(0.75)  #Q3 - 75th percentile

  IQR=q3-q1

  outliers = data_frame[((data<(q1-1.5*IQR))|(data>(q3+1.5*IQR)))]
  return outliers
```

```python
#Pandas truncates the displayed data for larger datasets.
pd.set_option('display.max_rows', 500)

pd.set_option('display.max_columns', 500)

pd.set_option('display.width', 150)
```

```python
#function to display the detected outliers
outliers = find_outliers_IQR(data_frame['Age'])

print("number of outliers: "+ str(len(outliers)))

outliers
```

```python
#Remove outliers by index
data_frame.drop(outliers.index, inplace=True)
```

```python
# Show missing values before imputation
print("Missing values before imputation:")
data_frame.isnull().sum()
```

```python
#Impute missing Age values with mean
Mean_Age = data_frame['Age'].mean()
data_frame['Age'].fillna(Mean_Age, inplace=True)
```

```python
#Encode Mortality_Status to numeric
data_frame['Mortality_Status'] = data_frame['Mortality_Status'].map({'Alive': 0 , 'Dead': 1})
```

```python
# Drop rows with invalid Mortality_Status
data_frame = data_frame.dropna(subset=['Mortality_Status'])
```

```python
#Encode Grade variable ordinally
label_encoder = preprocessing.LabelEncoder()

data_frame['T_Stage']= label_encoder.fit_transform(data_frame['T_Stage'])
data_frame['N_Stage']= label_encoder.fit_transform(data_frame['N_Stage'])
data_frame['Grade']= label_encoder.fit_transform(data_frame['Grade'])
data_frame['Estrogen_Status']= label_encoder.fit_transform(data_frame['Estrogen_Status'])
data_frame['Progesterone_Status']= label_encoder.fit_transform(data_frame['Progesterone_Status'])
data_frame['Sex']= label_encoder.fit_transform(data_frame['Sex'])
data_frame['A_Stage']= label_encoder.fit_transform(data_frame['A_Stage'])
```
