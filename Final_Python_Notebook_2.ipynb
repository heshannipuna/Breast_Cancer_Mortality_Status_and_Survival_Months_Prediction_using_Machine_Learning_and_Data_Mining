```python
#importing pandas library
import pandas as pd
```

```python
#loading the classification dataset
class_data = pd.read_csv('/content/classification.csv')
```

```python
#displaying first twelve rows
class_data.head(12)
```

```python
#prints the summary of class_data that includes the data type of each colum
class_data.info()
```

```python
# check dataset dimensions
print("Dataset dimensions:", class_data.shape)
```

```python
# Check for missing values
print("Missing values:")
class_data.isnull().sum()
```

```python
#configure pandas to display all rows/columns without truncation
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)
```

```python
# displaying first five rows before declaring x and y features
class_data.head(5)
```

```python
#Declare features (X) and target (y) for modeling
feature_cols = ['Age', 'Grade', 'Tumor_Size', 'Regional_Node_Examined', 'Reginol_Node_Positive', 'T_Stage', 'N_Stage', 'Estrogen_Status', 'Progesterone_Status', 'A_Stage', 'Sex']

X = class_data[feature_cols]
y = class_data['Mortality_Status']
```

```python
# showing feature columns of the dataset
print('feature columns:', feature_cols)
```

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize and apply scaling to features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#Import train_test_split to split class_data into training/test subsets
from sklearn.model_selection import train_test_split
```

```python
#Spliting data (70% training, 30% test) with repoducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Display train/test set sizes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
#Importing the Logistic Regression algorithm
from sklearn.linear_model import LogisticRegression
```

```python
#Initialize Logistic Regression model
logreg = LogisticRegression()
```

```python
#Train the model on training data
logreg.fit(X_train, y_train)
```

```python
#Predict on training data to check learning performance
y_pred_train=logreg.predict(X_train)
```

```python
#Predict on test data for real-world performance
y_pred=logreg.predict(X_test)
```

```python
# Predict on test data
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Accuracy
acc_default_lr = accuracy_score(y_test, y_pred)

# Recall (Dead class = 1)
recall_default_lr = recall_score(y_test, y_pred, pos_label=1)

# F1-Score (Dead class = 1)
f1_default_lr = f1_score(y_test, y_pred, pos_label=1)

# AUC-ROC Score
auc_default_lr = roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1])

# Confusion Matrix
cm_default_lr = confusion_matrix(y_test, y_pred)

# Print results
print(f"Default Logistic Regression - Accuracy: {acc_default_lr}")
print(f"Default Logistic Regression - Recall: {recall_default_lr}")
print(f"Default Logistic Regression - F1-Score: {f1_default_lr}")
print(f"Default Logistic Regression - AUC-ROC: {auc_default_lr}")
print(f"Confusion Matrix:\n{cm_default_lr}")
```

```python
#Import GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
# Logistic Regression - hyperparameter tuning
param_grid_lr = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2'], 'solver': ['liblinear']}
```

```python
# grid search with 5-fold cross-validation
grid_search_lr = GridSearchCV(LogisticRegression(), param_grid_lr, cv=5, scoring='f1')
```

```python
# Fit model to training data
grid_search_lr.fit(X_train, y_train)
```

```python
# Retrieve best model from gridsearch
best_lr_model = grid_search_lr.best_estimator_
print("Best LR Hyperparameters:", grid_search_lr.best_params_)
```

```python
# trainin final model with best parameters
best_lr_model.fit(X_train, y_train)
```

```python
# Generate predictions for training and test sets
y_pred_train_lr = best_lr_model.predict(X_train)

y_pred_lr = best_lr_model.predict(X_test)
```

```python
# importing libraris for evaluation metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import RocCurveDisplay
```

```python
# Calculate train accuracy scores for LR
print("Logistic Regression Train Accuracy:", accuracy_score(y_train, y_pred_train_lr))
```

```python
# calculate test accuracy scores for LR
print("Logistic Regression Test Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
# Generate and plot the confusion matrix
cm_lr = confusion_matrix(y_test, y_pred_lr, labels=best_lr_model.classes_)
disp_lr = ConfusionMatrixDisplay(cm_lr, display_labels=best_lr_model.classes_)

disp_lr.plot()
```

```python
# Generating classification- reports precision,recall,F1
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
# plot AUC-ROC curve
roc_lr = RocCurveDisplay.from_estimator(best_lr_model, X_test, y_test)
```

```python
# Values only evauation matrics  for the report
# Confusion Matrix - valus only
lr_cm = confusion_matrix(y_test, y_pred_lr)
print("Logistic Regression Confusion Matrix values:\n", lr_cm)

# Classification Report
print("\nLogistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))

# AUC-ROC Score
lr_probs = logreg.predict_proba(X_test)[:, 1]
lr_auc = roc_auc_score(y_test, lr_probs)
print("\nLogistic Regression AUC-ROC Score:", lr_auc)
```
